## 1. Tensors

A **Tensor** is a multi-dimensional array containing elements of a single data type (usually floats). 

Let's see how to create tensors in PyTorch:

````python
# 1. Create a tensor directly from a list
data = [[1, 2], [3, 4]]
t_data = torch.tensor(data)

# 2. Create from a NumPy array
np_array = np.array([[5, 6], [7, 8]])
t_np = torch.from_numpy(np_array)

# 3. Create with random values, ones, or zeros
t_rand = torch.rand(2, 3)     # uniform distribution [0, 1)
t_ones = torch.ones(2, 3)     # populated with 1.0
t_zeros = torch.zeros(2, 3)   # populated with 0.0

print("t_data:\n", t_data)
print("t_np:\n", t_np)
print("t_rand:\n", t_rand)
print("Shape of t_rand:", t_rand.shape)
print("Datatype of t_rand:", t_rand.dtype)
````

## 2. Tensor Operations & Reshaping

PyTorch supports over 100 tensor operations. They can run on GPUs at massive speeds.

````python
tensor = torch.tensor([[1, 2], [3, 4]], dtype=torch.float32)

# Element-wise multiplication
print("Element-wise multiplication (tensor * tensor):\n", tensor * tensor)

# Matrix multiplication (equivalent to np.dot)
print("Matrix multiplication (tensor @ tensor):\n", tensor @ tensor)

# In-place addition (methods ending with '_' are in-place)
tensor.add_(5)
print("After in-place add_:\n", tensor)
````

### Reshaping and Transposing
Changing the dimensions of a tensor is extremely common (e.g. flattening an image matrix into a vector before sending it to a neural network classification layer).

````python
x = torch.rand(4, 3)

# Use .view() or .reshape() to change shapes
y_flat = x.view(12)        # 1D vector of length 12
y_reshaped = x.reshape(2, 6) # 2D matrix of shape 2x6

print("x Shape:", x.shape)
print("y_flat Shape:", y_flat.shape)
print("y_reshaped Shape:", y_reshaped.shape)
````

## 3. Autograd: Automatic Differentiation

Training neural networks requires computing gradients of a loss function with respect to weights. PyTorch's **Autograd** engine tracks all operations on tensors with `requires_grad=True` and automatically constructs a backpropagation graph.

Let's test this with a simple function: 
$$y = 3x^2 + 5x$$
The derivative is:
$$\frac{dy}{dx} = 6x + 5$$

If $x = 2.0$, then $\frac{dy}{dx} = 6(2) + 5 = 17$.

````python
# Create a scalar tensor, setting requires_grad=True to track computations
x = torch.tensor(2.0, requires_grad=True)

# Define the function y = 3*x^2 + 5*x
y = 3 * (x ** 2) + 5 * x

print(f"Value of y: {y.item()}") # Item extracts scalar float

# Compute gradients (Backpropagation)
y.backward()

# Output the derivative dy/dx
print(f"Gradient of y with respect to x (dy/dx): {x.grad.item()}")
````

### Turning Off Gradient Tracking
When evaluating models or running inference, we do not need to track gradients. Disabling tracking saves memory and computation:

```python
with torch.no_grad():
    # operations here will not track gradients
```

## 4. CUDA and GPU Acceleration

If you have an NVIDIA GPU, you can drastically speed up tensor operations by moving them from CPU to GPU. 

The standard practice to write device-agnostic code is:

````python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Create a tensor on the CPU
x = torch.ones(2, 2)

# Move the tensor to the active device
x = x.to(device)
print("Tensor device:", x.device)
````
