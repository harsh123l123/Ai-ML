```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

---

## 1. Linear Regression Mathematics

### Hypothesis Function
We model the relationship between input $x$ and output $y$ as a straight line:
$$\hat{y} = w x + b$$
where:
- $w$ is the **slope** (weight).
- $b$ is the **y-intercept** (bias).

### Cost Function: Mean Squared Error (MSE)
We evaluate how wrong our model is by squaring the prediction errors and averaging them over all $m$ training samples:
$$J(w, b) = \frac{1}{2m} \sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)})^2 = \frac{1}{2m} \sum_{i=1}^m (w x^{(i)} + b - y^{(i)})^2$$

### Optimization: Gradient Descent
To minimize the cost $J$, we iteratively update $w$ and $b$ in the opposite direction of the gradients (slopes of the cost curve):
$$w \leftarrow w - \alpha \frac{\partial J}{\partial w}$$
$$b \leftarrow b - \alpha \frac{\partial J}{\partial b}$$

where $\alpha$ is the **learning rate** (step size). The partial derivatives are:
$$\frac{\partial J}{\partial w} = \frac{1}{m} \sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)}) x^{(i)}$$
$$\frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)})$$

## 2. Generating Synthetic Linear Data

Let's generate data points around the line $y = 3x + 5$ with added Gaussian noise.

```python
# Generate 100 random points
X = 2 * np.random.rand(100, 1)
y = 5 + 3 * X + np.random.randn(100, 1)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.7, label='Data points')
plt.xlabel('X (Feature)')
plt.ylabel('y (Target)')
plt.title('Synthetic Linear Dataset')
plt.grid(True)
plt.legend()
plt.show()
```

---

## 3. Implementing Linear Regression From Scratch

```python
class LinearRegressorScratch:
    def __init__(self, learning_rate=0.1, iterations=100):
        self.lr = learning_rate
        self.iterations = iterations
        self.w = 0.0
        self.b = 0.0
        self.cost_history = []
        
    def fit(self, X, y):
        m = len(X)
        # Reshape to 1D flat arrays for calculation simplicity
        x_flat = X.flatten()
        y_flat = y.flatten()
        
        for i in range(self.iterations):
            # Hypothesis predictions
            y_pred = self.w * x_flat + self.b
            
            # Calculate Cost (MSE)
            cost = (1 / (2 * m)) * np.sum((y_pred - y_flat) ** 2)
            self.cost_history.append(cost)
            
            # Gradients calculation
            dw = (1 / m) * np.sum((y_pred - y_flat) * x_flat)
            db = (1 / m) * np.sum(y_pred - y_flat)
            
            # Weights and bias updates
            self.w -= self.lr * dw
            self.b -= self.lr * db
            
            # Print progress every 10 iterations
            if i % 10 == 0:
                print(f"Iteration {i:02d}: Cost = {cost:.4f}, w = {self.w:.4f}, b = {self.b:.4f}")
```

```python
# Instantiating and training
model_scratch = LinearRegressorScratch(learning_rate=0.1, iterations=100)
model_scratch.fit(X, y)

print(f"\nFinal weight (w): {model_scratch.w:.4f} (True w = 3.0)")
print(f"Final bias (b): {model_scratch.b:.4f} (True b = 5.0)")
```

### Plotting Cost History and Final Fit

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))

# Cost reduction curve
ax1.plot(model_scratch.cost_history, color='purple', linewidth=2)
ax1.set_title("Cost History (Gradient Descent Convergence)")
ax1.set_xlabel("Iteration")
ax1.set_ylabel("MSE Cost")
ax1.grid(True)

# Final regression line overlay
ax2.scatter(X, y, color='blue', alpha=0.7, label='Data points')
x_range = np.linspace(0, 2, 100)
y_line = model_scratch.w * x_range + model_scratch.b
ax2.plot(x_range, y_line, color='red', linewidth=3, label=f'Fit Line (y = {model_scratch.w:.2f}x + {model_scratch.b:.2f})')
ax2.set_title("Linear Regression Fit Overlay")
ax2.set_xlabel("X")
ax2.set_ylabel("y")
ax2.grid(True)
ax2.legend()

plt.show()
```

## 4. Linear Regression using Scikit-Learn

In professional settings, we use optimized libraries like `scikit-learn` instead of writing algorithms from scratch. Let's see how simple it is.

```python
# Instantiate scikit-learn model
sk_model = LinearRegression()

# Fit model
sk_model.fit(X, y)

# Predict
y_pred = sk_model.predict(X)

# Coefficients
print(f"Scikit-Learn w: {sk_model.coef_[0][0]:.4f}")
print(f"Scikit-Learn b: {sk_model.intercept_[0]:.4f}")

# Evaluation
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared (Accuracy of fit): {r2:.4f}")
```
