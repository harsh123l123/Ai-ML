## 1. Vectors

A **vector** is an ordered list of numbers. In geometry, a vector represents a direction and a magnitude starting from the origin.

### Vector Operations
- **Addition**: $\vec{u} + \vec{v} = \begin{bmatrix} u_1 \\ u_2 \end{bmatrix} + \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} = \begin{bmatrix} u_1 + v_1 \\ u_2 + v_2 \end{bmatrix}$
- **Scalar Multiplication**: $c \vec{u} = \begin{bmatrix} c \cdot u_1 \\ c \cdot u_2 \end{bmatrix}$
- **Dot Product**: $\vec{u} \cdot \vec{v} = u_1 v_1 + u_2 v_2 = \sum u_i v_i$

```python
# Define two vectors
u = np.array([2, 3])
v = np.array([4, -1])

# Vector addition
u_plus_v = u + v

# Scalar multiplication
scaled_u = 2 * u

# Dot product
dot_product = np.dot(u, v) # or u @ v

print(f"u: {u}")
print(f"v: {v}")
print(f"u + v: {u_plus_v}")
print(f"2 * u: {scaled_u}")
print(f"u . v: {dot_product}")
```

### Geometric Visualization
Let's visualize vector addition geometrically. The sum of two vectors $\vec{u} + \vec{v}$ represents moving along $\vec{u}$ and then moving in the direction/distance of $\vec{v}$.

```python
plt.figure(figsize=(8, 8))

# Plot vectors using quiver
origin = np.array([0, 0])
plt.quiver(*origin, *u, color='r', angles='xy', scale_units='xy', scale=1, label='u (2, 3)')
plt.quiver(*origin, *v, color='b', angles='xy', scale_units='xy', scale=1, label='v (4, -1)')

# Plot u + v starting from u (tip-to-tail)
plt.quiver(*u, *v, color='g', alpha=0.5, angles='xy', scale_units='xy', scale=1, linestyle='--', label='v shifted')
plt.quiver(*origin, *u_plus_v, color='g', angles='xy', scale_units='xy', scale=1, label='u + v (6, 2)')

plt.xlim(-1, 8)
plt.ylim(-2, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both')
plt.legend()
plt.title("Geometric Vector Addition")
plt.show()
```

## 2. Matrices

A **matrix** is a 2D grid of numbers. In ML, a matrix can represent a dataset (where rows are samples, columns are features) or network weights.

### Matrix Multiplication
Multiplying matrix $A$ of shape $(m \times n)$ by matrix $B$ of shape $(n \times p)$ yields matrix $C$ of shape $(m \times p)$.
$$C_{ij} = \sum_{k=1}^n A_{ik} B_{kj}$$

*Note: Matrix multiplication is not commutative ($A B \neq B A$).*

```python
# Define two matrices
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

# Matrix addition
matrix_sum = A + B

# Element-wise multiplication (Hadamard product)
elementwise_mul = A * B

# Dot product / Matrix multiplication
matrix_mul = np.dot(A, B) # or A @ B

# Transpose of A
A_transpose = A.T

print("A:\n", A)
print("B:\n", B)
print("A + B:\n", matrix_sum)
print("Element-wise multiplication (A * B):\n", elementwise_mul)
print("Matrix Multiplication (A @ B):\n", matrix_mul)
print("Transpose of A:\n", A_transpose)
```

## 3. Solving Systems of Linear Equations

Many ML optimization problems boil down to solving linear systems $A \vec{x} = \vec{b}$, where $A$ and $\vec{b}$ are known, and we want to find $\vec{x}$.

If $A$ is invertible ($A^{-1}$ exists), then:
$$\vec{x} = A^{-1} \vec{b}$$

Let's solve:
$$\begin{cases} 2x + y = 5 \\ x - 3y = -1 \end{cases}$$
Which is:
$$\begin{bmatrix} 2 & 1 \\ 1 & -3 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 5 \\ -1 \end{bmatrix}$$

```python
A = np.array([
    [2, 1],
    [1, -3]
])
b = np.array([5, -1])

# Check determinant to see if invertible (det != 0)
det_A = np.linalg.det(A)
print(f"Determinant of A: {det_A:.2f}")

if det_A != 0:
    # Solve using solve (more numerically stable than direct inversion)
    x = np.linalg.solve(A, b)
    print(f"Solution x: {x}")
    
    # Verify solution: Ax = b
    print("Verify A @ x:", A @ x)
else:
    print("Matrix is singular (not invertible).")
```

## 4. Eigenvalues and Eigenvectors

For a square matrix $A$, a non-zero vector $\vec{v}$ is an **eigenvector** if multiplying by $A$ only scales the vector by a scalar $\lambda$, known as the **eigenvalue**:

$$A \vec{v} = \lambda \vec{v}$$

Eigenvectors represent directions of pure scaling under linear transformations, which makes them highly useful in dimensionality reduction techniques like Principal Component Analysis (PCA).

```python
# Define a symmetric matrix (symmetric matrices always have real eigenvalues)
A = np.array([
    [4, 2],
    [2, 3]
])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors (columns are eigenvectors):\n", eigenvectors)

# Let's verify: A @ v = lambda * v for the first eigenpair
lambda_0 = eigenvalues[0]
v_0 = eigenvectors[:, 0]

lhs = A @ v_0
rhs = lambda_0 * v_0

print(f"\nFirst Eigenvalue: {lambda_0:.4f}")
print(f"First Eigenvector: {v_0}")
print(f"A @ v_0: {lhs}")
print(f"lambda_0 * v_0: {rhs}")
print("Verification check:", np.allclose(lhs, rhs))
```
